```python
import itertools
import math
from gurobipy import Model, GRB

def solve_linear_program():
    # Create a Gurobi model
    model = Model("Linear_Programming_Example")
    
    k = 6
    
    Rs = {}
    for i in range(1, k//2 + 1):
        Rs[i] = [(j, i) for j in range(1, k + 1)]
    
    Ts = {}
    for i in range(1, k + 1):
        Ts[i] = [(i, j) for j in range(1, k//2 + 1)]

    variables = {}
    all_variables = set()
    for v in Rs.values():
        all_variables.update(v)
    for v in Ts.values():
        all_variables.update(v)
    
    #print(Rs)
    print(all_variables)
    
    for i in range(1, len(all_variables) + 1):
        for subset in itertools.combinations(all_variables, i):
            #print(subset)
            name = str(subset)
            x = model.addVar(lb=0, ub=math.log(len(subset), 2), name=name)
            variables[frozenset(subset)] = x
    
    model.setObjective(variables[frozenset(all_variables)], GRB.MAXIMIZE)

    # Add constraints
    for rel in Rs.values():
        model.addConstr(variables[frozenset(rel)] <= 1)
    
    for rel in Ts.values():
        model.addConstr(variables[frozenset(rel)] <= 1)
    
    
    for j in range(1, k//2 + 1):
        fd_vars = [(i, j) for i in range(1, k + 1)]
        for set_size in range(1, len(fd_vars) + 1):
            for subset in itertools.combinations(fd_vars, set_size):
                for k in range(1, k):
                    Y = frozenset(subset)
                    Yx = frozenset(list(subset) + [(k, j)])
                    model.addConstr(variables[Yx] - variables[Y] == 0)

    # Optimize the model
    model.optimize()

    # Check optimization status
    if model.status == GRB.OPTIMAL:
        print("Optimal solution found!")
        print(f"Objective value = {model.objVal}")
    elif model.status == GRB.INFEASIBLE:
        print("The model is infeasible.")
    elif model.status == GRB.UNBOUNDED:
        print("The model is unbounded.")
    else:
        print(f"Optimization ended with status {model.status}")

# Call the function
solve_linear_program()
```
